Experiment 2::study of basic python libraries
//executing some functions

```python
import math
```

```python
num1 = 9
num2 = 1000
angle = 30
```

```python
print("Square Root:", math.sqrt(num1))
print("Factorial:", math.factorial(num1))
print("Logarithm base 10:", math.log10(num2))
print("Sine of angle:", math.sin(math.radians(angle)))
print("Value of Pi:", math.pi)

```

//applying ceil functions

```python
num1 = -0.4
num2 = 6.9
num3 = -9.5
num4 = 4.99
```

```python
print("Ceil of", num1, "is:", math.ceil(num1))
print("Ceil of", num2, "is:", math.ceil(num2))
print("Ceil of", num3, "is:", math.ceil(num3))
print("Ceil of", num4, "is:", math.ceil(num4))
```
